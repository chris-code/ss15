# Linear Regression

This tutorial shows how simple it is to implement you own linear regression in python.

1) We need some data so create an array x with shape (100,1) that contains the numbers from -5 to 5.

2) Initialize the random number generator of numpy to 42.

3) Now add Gaussian random noise to x with a standard deviation of 1 and store the result in and arry y.

4) Plot the datapoints as yellow dots in the range $x\in[-10,10]$ and $y\in[-10,10]$

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

np.random.seed(42)
x = np.arange(100)*0.1-5.0
y = x+np.random.randn(100)

x = x.reshape(100,1)
y = y.reshape(100,1)

plt.plot(x,y, 'yo')
plt.xlim(-10, 10)
plt.ylim(-10, 10)
```

A linear regression for datapoint matrix $X$ ($D \times N$, D datapoints and N input dimensions) and target matrix $Y$ ($D \times M$, D datapoints and M output dimensions) is defined as:

$min \frac{1}{2}\left(A\vec{x}-\vec{y}\right)^2$

We ignore the bias value for now!

1) Set the derivative to zero and solve the equation for $A$ to get the optima $A^*$. (Hint: If you have problems with the closed form solution with abitray dimensions, first solve the equation for 1D input and output)

2) Now plot the resulting

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

np.random.seed(42)
x = np.arange(100)*0.1-5.0
y = x+np.random.randn(100)

x = x.reshape(100,1)
y = y.reshape(100,1)

A = np.dot(np.dot(np.linalg.inv(np.dot(x.T,x)),x.T),y)

plt.plot(x,y, 'yo', x, np.dot(x,A), '--k')
plt.xlim(-10, 10)
plt.ylim(-10, 10)
```

A commen way to integrate a bias value for many machine learning methods is to add a dimension which is constant one for all datapoints!

1) Modify the code by adding a second constant dimension to x and add 10 to y to shift the datapoints verctically.

2) Now plot the result in the range $x\in[-10,10]$ and $y\in[0,20]$, notice that you have to select the first dimension of x in order not to plot the constant dimension!

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

np.random.seed(42)
x = np.arange(100)*0.1-5.0
y = x+np.random.randn(100)

x = np.vstack((x,x*0+1)).T
y = y + 10

x = x.reshape(100,2)
y = y.reshape(100,1)

A = np.dot(np.dot(np.linalg.inv(np.dot(x.T,x)),x.T),y)

plt.plot(x[:,0].reshape(100,1),y, 'yo', x[:,0].reshape(100,1), np.dot(x,A), '--k')
plt.xlim(-10, 10)
plt.ylim(0, 20)
```

By using a polynomial expansion of x we can fit a polynome to the data.

Fit a ploynome of degree 5 to the data

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

np.random.seed(42)
x = np.arange(100)*0.1
x = x.reshape(100,1)

x = np.hstack((x,x**2,x**3,x**4,x**5,x*0+1))
y = np.cos(x[:,0])+np.random.randn(100)*0.5

x = x.reshape(100,6)
y = y.reshape(100,1)

A = np.dot(np.dot(np.linalg.inv(np.dot(x.T,x)),x.T),y)

plt.plot(x[:,0].reshape(100),y, 'yo', x[:,0].reshape(100), np.dot(x,A), '--k')
plt.xlim(-0, 10)
plt.ylim(-2, 2)
```

Now perform the same using the linear regression function np.polyfit(x,y,5) of numpy. Notice that x,y are here 1D arrays! 

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

np.random.seed(42)
x = np.arange(100)*0.1
y = np.cos(x)+np.random.randn(100).reshape(100)*0.5

fit = np.polyfit(x,y,5)
fit_fn = np.poly1d(fit) 

plt.plot(x,y, 'yo', x, fit_fn(x), '--k')
plt.xlim(0, 10)
plt.ylim(-2, 2)
```
